I denne oppgaven lager jeg en interaktiv graf med ønskede verdier.
Bruker et datasett fra høstsemesteret vi jobbet med i R men brukte den litt andreledes.

```python
import pandas as pd
#leser inn dataene
alfa = pd.read_csv("https://github.com/owid/co2-data/raw/master/owid-co2-data.csv", index_col=0)#reading data
alfa
```

```python
#resetter indexen
df=alfa.reset_index()
#velger de kolonnene jeg skal bruke
df = pd.DataFrame(alfa, columns= ['country','year','co2_per_capita','gdp'])

df
```

```python
#velger radene jeg skal ha som er Norge og Sverige
resultat = df.loc[["NOR", "SWE"]]

resultat
  
```

```python
#sjekker at jeg har alle dataene jeg skal bruke
resultat.keys()
```

```python
#gir farge til landene
colors={'Norway': '#E14827', 'Sweden': '#2792E1'}



resultat['colors']=resultat['country'].apply(lambda cont: colors[cont])
resultat
```

```python
#lager nye navn
land = resultat.rename(columns = {"country":"Land",
                                  "year":"År",
                                 "co2_per_capita" : "Co2_per_capita",
                                  "gdp": "Bnp"})

#sjekker on jeg har objekter eller faktiske tall
land.dtypes
```

```python
#fjerner manglende verdier for begge to
land=land.dropna()
```

```python
land
```

```python
import numpy as np
from matplotlib import pyplot as plt

fig,ax=plt.subplots()

#lager labels:
ax.set_ylabel('Co2_per_capita i tonn')
ax.set_xlabel('År')


#skriver hva vi skal ha i plottet og setter farge:
ax.scatter(land['År'], land['Co2_per_capita'], label=' Co2 utslipp i Norge og Sverige')
ax.legend(loc='upper left',frameon=False)
```

```python
#da jeg ikke fikk til med å få inn forskjellige farger prøver jeg å dele det i to dataframes..
Norge = land.loc[["NOR"]]
Norge
```

```python
Sverige = land.loc[["SWE"]]
Sverige
```

```python
fig,ax=plt.subplots()

#lager labels:
ax.set_ylabel('Co2_per_capita i tonn')
ax.set_xlabel('År')


#skriver hva vi skal ha i plottet og setter farge:
ax.scatter(Sverige['År'], Sverige['Co2_per_capita'], label='Co2 utslipp i Sverige')
ax.scatter(Norge['År'],Norge['Co2_per_capita'], color='red', label='Co2 utslipp i Norge')
ax.legend(loc='upper left',frameon=False)
```

```python
#slik ble grafen og vi ser at på et tidspunkt hadde sverige større utslipp enn norge per capita
#mens de nå i dag har lavere utslipp en norge.

#for å lage denne interaktive grafen starter jeg slik:
land['size']=land['Co2_per_capita'].apply(lambda Co2: Co2**1)
land
```

```python
land['Bnp']=land['Bnp'].apply(lambda Bnp: Bnp**0.5)
land
```

```python
from bokeh.plotting import figure

#lager figur:
a = figure(
        title = "Co2 utslipp i Norge og Sverige", 
        x_axis_label = 'År',
        y_axis_label = 'co2 utslipp',
    
        x_axis_type = "log",

    
        tools="hover", 
        tooltips = [
            ("Land","@Land"),
            ("Co2prcapita", "@Co2_per_capita"),
            ("BNP", "@Bnp")
            ],
    
        plot_height = 580,
        plot_width = 980)
```

```python
from bokeh.io import show, output_notebook
from bokeh.plotting import output_file

a.scatter(
    source=land,
    x= 'År', 
    y= 'Co2_per_capita',
    legend_field='Land',
    color='colors',
    
    size='size',
    alpha= 0.7
    )
a.legend.location='top_left'

#lager graf og lagrer den til html
output_notebook()
output_file("co2utslipp.html")
show(a)
```
